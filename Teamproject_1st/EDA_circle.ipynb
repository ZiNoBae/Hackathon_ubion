```python
import sweetviz
```

```python
# train & test 분리
df_train = df.loc[(df['연도'] >= 2016) & (df['연도'] < 2021)]
df_test = df.loc[(df['연도'] >= 2021) & (df['연도'] < 2023)]
```

```python
df_report = sweetviz.compare([df_train, 'Train'], [df_test, 'Test'], '물건금액(만원)')
```
