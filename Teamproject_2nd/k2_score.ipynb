```python
# 3. 누적수익성 
k2_score['누적수익성'] = df_altman['평균이익잉여금']/df_altman['평균총자산']
```

```python
col_order = ['회사명', '거래소코드', '회계년도','log총자산(백만원)','총자산회전율(log)', '누적수익성', '재무구조']
```

```python
k2_score = k2_score.reindex(columns = col_order)
k2_score
```

# K2_score 적용

```python
k2_score.columns
```

```python
k2_series = (-18.696) + 1.501*k2_score['log총자산(백만원)'] + 2.706*k2_score['총자산회전율(log)'] + 19.760*k2_score['누적수익성'] + 1.146*k2_score['재무구조'] 
```

```python
k2_score['K2_score'] = k2_series
```

```python
k2_score
```

```python
# K2 score 범위 확인 

## 부실 가능성 심각
bad = (k2_score['K2_score'] < (-2.3)).sum()

## 판정 보류
neutral = ((k2_score['K2_score'] >= -2.3) & (k2_score['K2_score']<=0.75)).sum()

## 건전판정
good = (k2_score['K2_score'] > 0.75).sum()

###
agg = bad + neutral + good

print(f'부실 비중 : {np.round(((bad/agg)*100),4)}%')
print(f'보류 비중 : {np.round(((neutral/agg)*100),4)}%')
print(f'건전 비중 : {np.round(((good/agg)*100),4)}%')
```
