- 컬럼 이름 재변경 & 단위 정리 

```python
# 총자산회전율
main_df['총자산회전율(%->배)'] = main_df['총자산회전율(%->배)']/100

# 총자산증가율
main_df['총자산_증가율(배->%)'] = main_df['총자산_증가율(배->%)']*100

# 외화포지션
main_df['외화포지션(배->%)'] = main_df['외화포지션(배->%)']*100 


# 컬럼이름 재변경
main_df.rename(columns={'총자산회전율(%->배)':'총자산회전율(배)',
                        '총자산_증가율(배->%)' : '총자산_증가율(%)',
                        '외화포지션(배->%)' : '외화포지션(%)'}, inplace=True)

main_df[['총자산회전율(배)', '총자산_증가율(%)', '외화포지션(%)']]
```
