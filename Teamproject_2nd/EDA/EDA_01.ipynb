```python
import pandas as pd
```

```python
df = pd.read_excel('./dataset_2nd/df_final.xlsx', index_col=0)
```

```python
import sweetviz
```

```python
df
```

```python
df.info()
```

```python
for i in range(0, len(df.index), 1) : 
    df['year'][i] = df['회계년도'][i][0:4]
```

```python
df['year'] = df['year'].astype(int)
```

```python
df_train = df.loc[(df['year'] >= 2011) & (df['year'] < 2020)]
df_test = df.loc[(df['year'] >= 2020) & (df['year'] < 2023)]
```

```python
df_test
```

```python
df_report = sweetviz.compare([df_train, 'Train'], [df_test, 'Test'])
```

```python
df_report.show_html('./EDA_circle.html')
```
