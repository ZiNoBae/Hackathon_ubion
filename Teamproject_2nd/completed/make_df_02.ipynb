## II-1. 평균계산 : df_finance_calc -> result_calc

```python
grouped_finance_calc = df_finance_calc.groupby("거래소코드")
```

```python
result_calc = pd.DataFrame()

for name, group in grouped_finance_calc:
    group['전년도총자산'] = group['[제조]자산(*)(IFRS연결)'].shift(1)
    group['평균총자산'] = (group['[제조]자산(*)(IFRS연결)'] + group['전년도총자산']) / 2

    group['전년도총부채'] = group['[제조]부채(*)(IFRS연결)'].shift(1)
    group['평균총부채'] = (group['[제조]부채(*)(IFRS연결)'] + group['전년도총부채']) / 2

    group['전년도이익잉여금'] = group['[제조]      이익잉여금(결손금)(*)(IFRS연결)'].shift(1)
    group['평균이익잉여금'] = (group['[제조]      이익잉여금(결손금)(*)(IFRS연결)'] + group['전년도이익잉여금']) / 2

    group['전년도총자본'] = group['[제조]자본(*)(IFRS연결)'].shift(1)
    group['평균총자본'] = (group['[제조]자본(*)(IFRS연결)'] + group['전년도총자본']) / 2

    grouped_calc = group[['회사명','거래소코드','회계년도','평균총자산','평균총부채','평균이익잉여금','평균총자본','[제조]매출액(수익)(*)(IFRS연결)']]
    grouped_calc = grouped_calc.reset_index()
    result_calc = result_calc.append(grouped_calc)
```

## II-3. result_calc + 계산 컬럼 추가

```python
result_calc["청산가치율"] = result_calc["평균총자본"] / result_calc["[제조]매출액(수익)(*)(IFRS연결)"]*100
```

```python
result_calc["이익잉여금비율"] = result_calc["평균이익잉여금"] / result_calc["평균총자본"]*100
```

```python
result_calc["총자산회전율"] = result_calc["[제조]매출액(수익)(*)(IFRS연결)"] / result_calc["평균총자산"]*100
```

```python
result_calc["총자산_증가율"] = (result_calc["평균총자산"] - result_calc["평균총자산"].shift(1)) / result_calc["평균총자산"].shift(1)
```

```python
# result_calc.loc[result_calc['회계년도'].str.contains('2011'),['총자산_증가율']]
result_calc
```

```python
# 총자산 증가율 Pivot 
# 각 기업 첫줄 data 뽑고(1819행) -> 0으로 대체 

# result_calc.loc[result_calc.groupby(['회사명', '거래소코드'])[["총자산_증가율"]].apply(lambda x: x.iloc[0]),"총자산_증가율"]
for value in result_calc.groupby(['회사명', '거래소코드'])[['회계년도',"총자산_증가율"]].apply(lambda x: x.iloc[0]).reset_index().values.tolist():
    result_calc.loc[(result_calc["회사명"]==value[0]) &\
            (result_calc["거래소코드"]==value[1]) &\
                (result_calc["회계년도"]==value[2]) &\
                    ((result_calc['총자산_증가율'].isna())|(result_calc['총자산_증가율']==value[3])), '총자산_증가율'] = 0
    

result_calc.groupby(['회사명', '거래소코드'], as_index=False).apply(lambda x: x.iloc[0])



# result_calc.groupby(['회사명', '거래소코드'], as_index=False).apply(lambda x: x.iloc[0])
```
