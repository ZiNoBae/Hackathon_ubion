```python
df0 = pd.concat(
    [df[['회사명', '거래소코드', '회계년도','평균총자산', '평균총부채', '평균총자본','평균이익잉여금','[제조]매출액(수익)(*)(IFRS연결)']],
     df_mk[['시가총액']]], axis=1)

df0
```

# I. K2-Score 컬럼 확보

```python
k2_score = df0.iloc[:, 0:3]
k2_score
```

## I-1. 기업규모 : log(총자산)
    - 단위 : 천원 -> 원

```python
# 천원 ->원
k2_score['log총자산(원)'] = np.log10(df0['평균총자산']*1000)
```

## I-2. 총자산회전율 : log(매출액 / 평균총자산)

```python
k2_score['총자산회전율(log)'] = np.log10(df0['[제조]매출액(수익)(*)(IFRS연결)']/df0['평균총자산'])
```

```python
# x 
df0.loc[k2_score['총자산회전율(log)'] == -(np.inf)][['회사명', '거래소코드', '회계년도', '[제조]매출액(수익)(*)(IFRS연결)','평균총자산']]
# df0.iloc[10353]
# np.log(0.003152)
```

## I-3. 누적수익성 : 평균이익잉여금 / 평균총자산


```python
k2_score['누적수익성'] = df0['평균이익잉여금']/df0['평균총자산']
```

## I-4. 재무구조 : 시가총액 / 평균총부채
    - 시가총액 단위 변경 : 1,000,000-> 1,000 for 부채와 단위 통일

```python
k2_score['재무구조'] = (df0['시가총액']*1000) / df0['평균총부채']
```

```python
k2_score
```

# II. K2-Score 계산 


```python
k2_series = (-18.696) + 1.501*k2_score['log총자산(원)'] + 2.706*k2_score['총자산회전율(log)'] + 19.760*k2_score['누적수익성'] + 1.146*k2_score['재무구조'] 
```

```python
k2_score['K2_score'] = k2_series
```

```python
k2_score
```

```python
# x
# (-18.696) + 1.501*k2_score['log총자산(백만원)'] + 2.706*k2_score['총자산회전율(log)'] + 19.760*k2_score['누적수익성'] + 1.146*k2_score['재무구조'] 
# (-18.696) + 1.501*11.591514+ 2.706*(-5.7597) + 19.760*(-0.564895) + 1.146*(12.318762)
# k2_score[k2_score['거래소코드'] == 115180]
```

```python
# K2 score 범위 확인 

## 부실 가능성 심각
bad = (k2_score['K2_score'] < (-2.3)).sum()

## 판정 보류
neutral = ((k2_score['K2_score'] >= -2.3) & (k2_score['K2_score']<=0.75)).sum()

## 건전판정
good = (k2_score['K2_score'] > 0.75).sum()

###
agg = bad + neutral + good

print(f'부실 비중 : {np.round(((bad/agg)*100),4)}%')
print(f'보류 비중 : {np.round(((neutral/agg)*100),4)}%')
print(f'건전 비중 : {np.round(((good/agg)*100),4)}%')
```
